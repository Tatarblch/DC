Python — это высокоуровневый интерпретируемый кроссплатформенный язык программирования.


Язык программирования — это строго определённый набор правил, благодаря которому возможно взаимодействие человека с компьютером. Иначе говоря, язык программирования — это инструмент, с помощью которого вы можете управлять компьютером.

Машинный код — (ещё его называют машинный язык) — это набор команд, которые непосредственно выполняются центральным процессором компьютера.

Транслятор — это программа, которая выполняет преобразование программы, написанной на высокоуровневом языке программирования, в программу, состоящую из машинных команд.


Компилятор — выполняет преобразование (компиляцию) программы, написанной на языке высокого уровня (исходный язык), в машинный код

Кроссплатформенность, или платформонезависимость — это способность программы работать более чем на одной аппаратной платформе или в более чем одной операционной системе.

Синтаксис языка программирования — это набор правил, на основе которых строится программа.
