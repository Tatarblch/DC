**Переменная** — это именованная область памяти компьютера, адрес которой позволяет получить доступ к данным (это общее определение из программирования, конкретная трактовка которого может меняться от языка к языку).

В самом простом понимании **память** —  это упорядоченная последовательность ячеек (байт), в которых располагаются данные. Каждая ячейка имеет свой порядковый номер — **адрес**.

Существует **два главных правила именования переменных**:

Название переменной должно состоять только из букв, цифр и знаков подчёркивания _.
Название переменной не должно начинаться с цифры.


→ Пожалуй, главный принцип, который стоит иметь ввиду при именовании переменных и в целом при написании кода: **«код читается чаще, чем пишется»**.


Более формально, **функция** — фрагмент кода, к которому можно обратиться из любого другого места.

**Алгоритм** — набор последовательных действий, направленных на достижение поставленной цели или решения конкретной задачи.

В программировании результат, который возвращает функция, называют **возвращаемым значением функции**.
